Loading Dataset

```python
from pandas import read_csv
```

```python
fileName = "/content/Energy Meter.csv"
names = ['Voltage', 'Current', 'Power', 'class']
dataset = read_csv(fileName, names=names)
```

Summarization

```python
print(dataset.shape)
print(dataset.head(20))
print(dataset.describe())
print(dataset.groupby('class').size())
```

Visualization

```python
from pandas.plotting import scatter_matrix
from matplotlib import pyplot

dataset.plot(kind = 'bar', subplots = True, layout = (2,2) )
pyplot.title('Bar Plot')
pyplot.show()

dataset.hist()
pyplot.title("Histogram plot")
pyplot.show()

scatter_matrix(dataset)
pyplot.title('Scattered plot')
pyplot.show()
```

ML Algorithms

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

```python
arr = dataset.values
X = arr[:, 0:3]
y = arr[:,3]
X_train, X_Validation, y_train, y_Validation = train_test_split(X,y, train_size=0.20, random_state=1, shuffle=True)
```

```python

models = []
models.append(('LR', LogisticRegression(solver='liblinear', multi_class='ovr')))
models.append(( 'LDA', LinearDiscriminantAnalysis()))
models.append(( 'KNN', KNeighborsClassifier(n_neighbors=5)))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC(gamma='auto')))
```

```python
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)

results = []
names = []
res = []

for name, model in models:

  kfold = StratifiedKFold(n_splits = 10, random_state=None)
  cv_results = cross_val_score(model, X_train, y_train, cv = kfold, scoring = 'accuracy')
  results.append(cv_results)
  names.append(name)
  res.append(cv_results.mean())
  print(' %s: %f (%f)' % (name, cv_results.mean(), cv_results.std()))


pyplot.ylim(.990,.999)
pyplot.bar( names, res, color = 'green', width = 0.6)

pyplot.title("Algorihtm Comparison")
pyplot.show()
```

Training & Testing

```python
model = SVC(gamma='auto')
model.fit(X_train, y_train)
```

```python
import pickle
file = '/content/finalized_model.pkl'
pickle.dump(model, open(file, 'wb'))
```

```python
loaded_model = pickle.load(open(file, 'rb'))
result = loaded_model.score(X_Validation, y_Validation)
print(result)
```
